```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '7'
from examples.ExVideo.ExVideo_svd_train import *
from diffsynth.models.svd_unet import SVDUNet, PushBlock, PopBlock, PopMixBlock
from avcv.all import *
from fastcore.all import patch

from diffsynth import save_video, ModelManager, SVDVideoPipeline, HunyuanDiTImagePipeline
from diffsynth import ModelManager
import torch, os

```

```python
args = parse_args().parse_known_args()[0]
args.num_frames = 4
args.steps_per_epoch = 500
args.pretrained_path =  "models/stable_video_diffusion/svd_xt.safetensors"
args.height = 512
args.width = 512
ds_path = './datasets/tiktokdance/'
steps_per_epoch = 500
dataset = TextVideoDataset(
        ds_path,
        os.path.join(ds_path, "metadata.json"),
        training_shapes=[(args.num_frames, 1, args.num_frames, 512, 384)],
        steps_per_epoch=args.steps_per_epoch)
```

```python
if not 'pipe' in dir():
    model_manager = ModelManager()
    model_manager.load_models(
        [
         "models/stable_video_diffusion/svd_xt.safetensors",
         "/anhvth5/vision-projects/DiffSynth-Studio/outputs/models_tiktokdance/lightning_logs/version_31/checkpoints/epoch=6-step=1995-ema.safetensors"
        ]
    )
    pipe = SVDVideoPipeline.from_model_manager(model_manager)
```

```python
pipe.unet.conv_in
```

```python
data = dataset[0]
```

```python
first_frame = dataset.raw_frames[0]
first_frame = first_frame[0].permute([1,2,0]).cpu().numpy().astype('uint8')
first_frame = Image.fromarray(first_frame).resize((384, 512))
first_frame
```

```python
frames = data['frames_0'].cuda().float()
```

```python
frames.shape
```

```python
with torch.cuda.amp.autocast(), torch.no_grad():
    torch.manual_seed(1)
    video = pipe.__call__(
        input_image=first_frame,
        num_frames=args.num_frames, fps=30, height=512, width=384,
        motion_bucket_id=127,
        num_inference_steps=50,
        min_cfg_scale=2, max_cfg_scale=2, contrast_enhance_scale=1.2,
        motion_frames = frames
    )
save_video(video, "video.mp4", fps=30)
```
